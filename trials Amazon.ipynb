```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
results=soup.find_all('a',{'class':"a-link-normal a-text-normal"})
dict_links={}

for result in results:
    link='https://www.amazon.es'+str(result['href'])
    descr=result.text
    dict_links[descr] = link

# dict_links
```

```python
props_dict={}
for item,link in list(dict_links.items())[:2]:
    driver.get(link)
    soup_result = BeautifulSoup(driver.page_source, 'html.parser')
    tech_table=soup_result.find('table',{'id':"productDetails_techSpec_section_1"})
    trs=tech_table.find_all('tr')   
    props_list=[]
    for tr in trs:
        th=tr.find('th').text
        td=tr.find('td').text
        props_list.append((th,td))
    props_dict[item]=props_list
     
    #Sacar ASIN, valoraciones, nota
        
props_dict
```

```python
url_bm = 'https://www.cpubenchmark.net/laptop.html'
html_bm = requests.get(url_bm).content
# print(html[:500])
soup_bm = BeautifulSoup(html_bm, 'html.parser')
# print(soup_bm)
table_bm=soup_bm.find('ul',{'class':"chartlist"})
```

```python
lis=table_bm.find_all('li')
cpu_dict={}
for li in lis:
    cpu_name=li.find('span',{'class':"prdname"}).text
    cpu_score=li.find('span',{'class':"count"}).text
    cpu_dict[cpu_name]=cpu_score
    
cpu_dict
```

```python
url_gpu = 'https://www.videocardbenchmark.net/gpu_list.php'
html_gpu = requests.get(url_gpu).content
# print(html[:500])
soup_gpu = BeautifulSoup(html_gpu, 'html.parser')
# print(soup_bm)
table_gpu=soup_gpu.find('table',{'id':"cputable"})
# table_gpu
```

```python
trs=table_gpu.find_all('tr')
gpu_dict={}
for tr in trs:
    tds=tr.find_all('td') 
    if tds!=[]:
        gpu_dict[tds[0].text]=tds[1].text
    
gpu_dict
```
